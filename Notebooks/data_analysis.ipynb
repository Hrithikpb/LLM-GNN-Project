```python
data = pd.read_csv('/Users/akshada/Desktop/akshada/LLM-GNN/LLMGNN/data/processed.csv')
data.head()
```

```python
data['title'] = data['title'].str.split(' ')
data['abstract'] = data['abstract'].str.split(' ')
```

```python
data["Lengths_Title"] = data.apply(lambda row: len(row.title), axis = 1)
data["Lengths_Abstract"] = data.apply(lambda row: len(row.abstract), axis = 1)
```
